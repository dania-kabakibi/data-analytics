# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
# Splits the code at the colon to create a list ["supplier", "productNum-size"], 
# then returns the first element using (index 0), which is the supplier code
def get_supplier_code(code):
    return code.split(":")[0]
```

```python
# First splits the code to isolate the product and size section ["supplier", "productNum-size"] using (index 1),
# then splits again by the hyphen ["productNum", "size"] to return the product number (index 0).
def get_product_num(code):
    productNum_size = code.split(":")[1]
    return productNum_size.split("-")[0]
```

```python
# Follows the same logic as get_product_num but returns the second element from the final split, which represents the size.
def get_size(code):
    productNum_size = code.split(":")[1]
    return productNum_size.split("-")[1]
```

```python
code1 = "ACME:123-L"
supplier_code1 = get_supplier_code(code1)
product_num1 = get_product_num(code1)
size1 = get_size(code1)

print(
    f"For {code1}, the supplier is {supplier_code1}. Part number {product_num1} is available in size {size1}."
)
```
