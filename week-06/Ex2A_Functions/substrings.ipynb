## Lab 1. Working with names – parsing strings

```python
full_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range(len(name)):
        if name[i] == ' ':
            first = name[:i]
            last = name[i+1:]
            print(f"Full name:  {name}")
            print(f"First name: {first}")
            print(f"Last name:  {last}")
```

```python
parse_and_display(full_name)
```

```python
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```python
parse_and_display(test_list[0]) # output for Grace Flores
```

```python
parse_and_display(test_list[1]) # output for JB Oinonen
```

```python
parse_and_display(test_list[2]) # output for Ken Penn
```

```python
parse_and_display('Dania Kabakibi')
```

```python
def parse_and_display2(name):
    spaces_index = []

    for i in range(len(name)):
        if name[i] == ' ':
            spaces_index.append(i)

    if len(spaces_index) == 2:
        first_name = name[:spaces_index[0]]
        middle_name = name[spaces_index[0] + 1:spaces_index[1]]
        last_name = name[spaces_index[1] + 1:]
        print(f"First name:  {first_name}")
        print(f"Middle name: {middle_name}")
        print(f"Last name:   {last_name}")

    elif len(spaces_index) == 1:
        first_name = name[:spaces_index[0]]
        last_name = name[spaces_index[0] + 1:]
        print(f"First name:  {first_name}")
        print(f"Last name:   {last_name}")

    else: 
        first_name = name
        print(f"First name:  {first_name}")
# Note: Using split() would make this function shorter and easier, but this version practices string slicing and indexing.
```

```python
parse_and_display2("Lorde")
```
