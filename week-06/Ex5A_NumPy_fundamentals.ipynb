```python
import numpy as np
```

## Array Basics

```python
fruits = ["apple", "banana", "cherry", "date"]
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(arr_1.dtype) # output: <U6
# U (Unicode string), 6 (The maximum length of the string in characters)
```

```python
arr_2 = np.array([[10, 11, 12], [13, 14, 15]])
print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)
# arr_2.T does not alter the value stored in arr_2
```

```python
arr_3 = np.array([[1, 3, 5], [2, 4, 6]])
print(arr_3)
```

```python
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```python
arr_5 = np.array([[[1, 1], [1, 1], [1, 1]]])
print(arr_5)
```
