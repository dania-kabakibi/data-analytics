# Exercise 2.C Creating and calling functions
## Lab 1. working with defining custom functions.

```python
def favorite_things(name, movie, music):
    return f"Name: {name}\n" f"Favorite movie: {movie}\n" f"Favorite musician / band: {music}"

print(favorite_things("Dania", "Inside out", "Musician"))
```

```python
def why_im_here(purpose):
    return purpose

print(why_im_here("Because I want to gain experince and get an internship"))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place("Amsterdam", "It has wonderful tulip fileds"))
```

```python
print(favorite_place("Flaming Grill (Chinese restaurant)", "I like Chinese food especially hibachi"))
```
