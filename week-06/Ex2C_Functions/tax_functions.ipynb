## Lab 3. Tax Calculation Functions
This lab demonstrates the creation of custom Python functions to automate payroll tax calculations. It focuses on:
- **Function Definition:** Creating reusable blocks of code with parameters and return values.
- **Conditional Logic:** Using `if/elif/else` statements to apply different tax rates based on withholding codes.
- **Mathematical Operations:** Implementing specific tax formulas, including complex logic for withholding adjustments.
- **Formatting:** Using f-strings to display financial data clearly.

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * 0.062 # tax rate of 6.2%
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145 # Medicare tax at 1.45%
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        tax_rate = 0.23
    elif withholding_code == 1:
        tax_rate = 0.21
    elif withholding_code == 2:
        tax_rate = 0.195
    elif withholding_code == 3:
        tax_rate = 0.185
    else:
        over_four = withholding_code - 4
        tax_rate = 0.18 - (over_four * 0.005)
    return gross_pay * tax_rate
```

```python
# Test the functions (Person1: gross pay $750 withholding code 0)
print("--- (Person1: gross pay $750 withholding code 0) ---")
print(
    f"The Social Security tax     : {get_soc_sec_tax(750):.2f}\n"
    f"The Medicare tax            : {get_medicare_tax(750):.2f}\n"
    f"The federal tax withholding : {get_federal_tax(750, 0):.2f}\n"
)
```
