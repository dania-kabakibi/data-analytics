## Lab 2. Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
day = today.strftime("%A") # Full weekday name
month = today.strftime("%B") # Month name
day_of_month = today.strftime("%d") # Day dd
year = today.strftime("%Y") # Year yyyy

print(f"Today is {day}, {month} {day_of_month}, {year}")
```

# My birthday

```python
my_birthday = datetime.date(1999, 6, 15)
```

```python
birthday_month = my_birthday.strftime("%m")
birthday_day = my_birthday.strftime("%d")
birthday_year = my_birthday.strftime("%Y")

print(f"My birthday is {birthday_month}/{birthday_day}/{birthday_year}")
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
ninety_m = ninety_d.strftime("%B")
ninety_dom = ninety_d.strftime("%d")
ninety_y = ninety_d.strftime("%Y")

print(f"90 days from today is {ninety_m} {ninety_dom}, {ninety_y}")
```

# Time for Dinner

```python
dinner_time = datetime.time(7, 30)
dinner_h = dinner_time.strftime("%I")
dinner_m = dinner_time.strftime("%M")

print(f"Let’s meet for dinner at {dinner_h}:{dinner_m} PM")
```
