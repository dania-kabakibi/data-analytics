### Exercise 2.E Using functions for analyzing data 
This notebook analyzes a small sales dataset using Python functions. 
Each function takes the full dataset as its first argument.

```python
sales_data = [
    ('Laptop',     'East',  1200.00),
    ('Phone',      'West',   850.00),
    ('Tablet',     'East',   400.00),
    ('Monitor',    'South',  650.00),
    ('Keyboard',   'West',    75.00),
    ('Laptop',     'South', 1150.00),
    ('Phone',      'East',   900.00),
    ('Tablet',     'West',   425.00),
    ('Monitor',    'East',   680.00),
    ('Keyboard',   'South',   80.00),
    ('Laptop',     'West',  1300.00),
    ('Phone',      'West',  820.00),
]
```

## Lab 1. Total sales by region

```python
def region_total(data, region):
    return sum(amt for _, reg, amt in data if reg == region)
```

```python
print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```python
def region_count(data, region):
    return sum(1 for _, reg, _ in data if reg == region)
```

```python
print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))
```

## Lab 3. Average sale amount by region

```python
def region_average(data, region):
    avg = 0
    if region_count(data, region) > 0:
        avg = region_total(data, region) / region_count(data, region)
    return avg
```

```python
print(f"{region_average(sales_data, 'East'):.2f}")
print(f"{region_average(sales_data, 'West'):.2f}")
print(f"{region_average(sales_data, 'South'):.2f}")
print(region_average(sales_data, 'North'))
```

## Lab 4. Filter by sale amount

```python
def sales_above(data, threshold):
    return (amt for _, _, amt in data if amt > threshold)
```

```python
for i in sales_above(sales_data, 500):
    print(i)
```

```python
def sales_below(data, threshold):
    return (amt for _, _, amt in data if amt < threshold)
```

```python
for i in sales_below(sales_data, 500):
    print(i)
```

## Lab 5. Top and bottom sale

```python
def top_sale(data):
    return max(data, key=lambda s: s[2])

def bottom_sale(data):
    return min(data, key=lambda s: s[2])

print(top_sale(sales_data))
print(bottom_sale(sales_data))
```

- The top sale was a Laptop in the West region for $1300.
- The bottom sale was a Keyboard in the West region for $75.

## Lab 6. Apply a discount

```python
def apply_discount(data, discount_pct):
    return list(map(lambda s: (s[0], s[1], s[2] * (1 - discount_pct)), data))

sales_data_d = apply_discount(sales_data, 0.15)

print("--- Before discount ---")
print(f"{sales_data[0]}\n{sales_data[1]}\n{sales_data[2]}\n")
print("--- After 15% discount ---")
print(f"{sales_data_d[0]}\n{sales_data_d[1]}\n{sales_data_d[2]}\n")
```

## Lab 7. Unique regions

```python
def unique_regions(data):
    return sorted(set(s[1] for s in data))

regions = unique_regions(sales_data)
print(f"Unique Regions: {regions}")
```

Extracting a sorted list of unique regions is a fundamental step in data analysis, because this list could be used to generate a dropdown menu, allowing a user to filter large amount of data.

## Lab 8. Region summary

```python
def region_summary(data, region):
    region_data = [s for s in data if s[1] == region]

    total = region_total(region_data, region)
    count = region_count(region_data, region)
    avg = region_average(region_data, region)
    top = top_sale(region_data)

    print(f"=== {region} Region ===")
    print(f"Records:        {count}")
    print(f"Total Sales:    ${total:,.2f}")
    print(f"Average Sale:   ${avg:,.2f}")
    print(f"Top Sale:       {top[0]} — ${top[2]:,.2f}")

region_summary(sales_data, "East")
```

## Lab 9. Full summary

```python
for r in unique_regions(sales_data):
    region_summary(sales_data, r)
    print()
```

In this series of labs, I learned how to build a modular data analysis in Python. Key takeaways include:

- Using generator expressions and set() to dynamically identify unique categories within a dataset.

- Implementing loops and higher-order functions like map and lambda to process data without repetitive hardcoding.

- Applying f-string formatting to present raw numerical data as professional, readable currency.
