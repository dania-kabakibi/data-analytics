# Exercise 2.B Cleaning up data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('test_data.csv')
```

```python
df.describe()
```

## Filtering a DataFrame

```python
df[(df['Age'] > 18)]
```

```python
df[(df['Score'] > 50) & (df['Score'] < 100)]
```

## Modifying a DataFrame

```python
df_cleaned = df.drop(df[df['Score'] < 0].index)
df_cleaned
```

```python
df_cleaned.drop('City', axis=1, inplace=True)
df_cleaned
```

```python
mean_age = np.mean(df['Age'])
df_cleaned['Age'] = df_cleaned['Age'].fillna(mean_age)
df_cleaned
```

```python
mean_score = np.mean(df['Score'])
df_cleaned.Score.replace(0, mean_score)
df_cleaned
```

```python
df_cleaned.dropna(inplace=True)
df_cleaned
```

```python
df_cleaned.index = df_cleaned.ID
df_cleaned.drop('ID', axis=1, inplace=True)
df_cleaned
```
