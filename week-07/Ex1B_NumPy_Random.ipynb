## Generating Random Choices from an Array

```python
students = np.array(["Aspenne", "Chantal", "Cris", "Demy", "Dimitri", "Hamzah", 
                     "Ivana", "Jonathan", "Jordan", "Lily", "Malaika", "Marenza"])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)

# This is not an effective way to assign breakout groups because np.random.choice 
# samples with replacement by default. This causes duplicate student assignments 
# across different groups and redundant names within the same group.
```

```python
call_on_weighted = np.random.choice(
    students, 
    p=[0.3, 0.0, 0.1, 0.1, 0.1, 0.1, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05],
    size=5
)
print(call_on_weighted)

# ["Aspenne", "Chantal", "Cris", "Demy", "Dimitri", "Hamzah", "Ivana", "Jonathan", "Jordan", "Lily", "Malaika", "Marenza"]
# [ 0.3,       0.0,       0.1,    0.1,    0.1,       0.1,      0.05,    0.05,       0.05,     0.05,   0.05,      0.05]
# I assigned a 0.3 probability to the first student and 0.0 to the second to satisfy 
# the prompt requirements. The remaining 0.7 probability was distributed among 
# the other students to ensure the total probability sums exactly to 1.0
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
