## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

I found that shuffle changed the order of the rows, while keeping the elements within each row in their original sequence.

```python
new_shuffle = np.random.permutation(r_4)
print("Using permutation():\n", new_shuffle)
print("\nThe Original array:\n", r_4)
```

I will use shuffle() when I want to modify the original array in place, and permutation() when I need a shuffled copy while leaving the original array unchanged.

## Grocery Problem

```python
# [Vegetables, Fruits, Snacks, Beverages]
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print("bag1:\n", bag1)
print("bag2:\n", bag2)
```

```python
bag1_total = np.sum(bag1)
bag2_total = np.sum(bag2)
print(f"The first bag contains {bag1_total} items, and the second bag contains {bag2_total} items")
```

```python
total = np.sum([bag1, bag2], axis=0)
print(f"There are {total[0]} vegetables, {total[1]} fruits, {total[2]} boxes of snacks, and {total[3]} beverages in the two grocery bags")
```

## Teamwork

```python
# [Mon, Tue, Wed, Thu, Fri]
ali = np.array([5, 4, 6, 4, 2])
```

```python
# [Mon, Tue, Wed, Thu, Fri]
ben = np.array([0, 5, 5, 3, 4])
```

```python
# [Mon, Tue, Wed, Thu, Fri]
carla = np.array([3, 6, 0, 3, 4])
```

```python
all_knocks = np.array([ali, ben, carla])
print("All door-knocks for the week:\n", all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1) # [21 17 16]
collective_total = np.sum(all_knocks)
print(f"Total doors knocked by person - Ali: {person_totals[0]}, Ben: {person_totals[1]}, Carla: {person_totals[2]}")
print(f"Total doors knocked by the group collectively: {collective_total}")
```

```python
daily_group_knocks = np.sum(all_knocks, axis=0) # [ 8 15 11 10 10]
rolling_knocks = np.cumsum(daily_group_knocks) # [ 8 23 34 44 54]

# Convert knocks to estimated votes (1 vote per 2 knocks)
estimated_votes = rolling_knocks / 2

print("Rolling total of estimated votes (Mon through Fri):", estimated_votes)
```
