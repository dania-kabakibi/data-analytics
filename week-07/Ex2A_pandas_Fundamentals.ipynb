```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
# ac = {9, 10, 11, 12}
ac = (9, 10, 11, 12)
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
# ser3 = pd.Series(ac) # TypeError: 'set' type is unordered
ser3 = pd.Series(ac)
ser4 = pd.Series(ad)

print("ser1:\n", ser1)
print("\nser2:\n", ser2)
# print("\nser3:\n", ser3)
print("\nser3:\n", ser3)
print("\nser4:\n", ser4)
```

```python
# ser5 = pd.Series(17, 18, 19, 20)
# TypeError: Index(...) must be called with a collection of some kind, 18 was passed

ser5 = pd.Series((17, 18, 19, 20))
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 1, ser2)
df1.insert(2, 2, ser3)
df1
```

```python
df1.insert(3, 3, ser4)
df1
# The series ser4 was created from a dictionary, meaning it contains a string-based 
# (named) index rather than a default integer index. As a result, it does not align 
# with the existing integer index of df1, producing NaN values.
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```

## Reading Data into a DataFrame

```python
df2 = pd.read_csv('sakila_example.csv')
df2
```

```python
df2.head() # 5 by default
```

```python
df2.tail(10)
```

```python
df2.info()
```

The .info() method returns the metadata for df2, showing that it is a DataFrame object. The RangeIndex displays the total number of rows along with the starting and ending indexes. It also provides the total number of columns, column names, non-null counts, and data types.

```python
df2.describe()
```

The .describe() method typically returns summary statistics for numeric data. However, because df2 consists of text columns, it instead returned the count, number of unique values, the top (most frequent) value, and its frequency. This information is useful for analyzing categorical trends and determining the total number of unique actors.

```python
df3 = pd.read_csv('northwind_example.csv')
df3
```

```python
df3.describe()
```
