# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
plt.plot(x, y, marker = 'o')

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Styling Plots

```python
plt.plot(x, y, 'o--r') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, 's--r', ms = 10, mfc = 'g') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, '^--r', ms = 10, mfc = 'c') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, 'D--b', ms = 10, mfc = 'm') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```
