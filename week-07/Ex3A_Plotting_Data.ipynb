# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
plt.plot(x, y, marker = 'o')

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Styling Plots

```python
plt.plot(x, y, 'o--r') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, 's--r', ms = 10, mfc = 'g') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, '^--r', ms = 10, mfc = 'c') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
plt.plot(x, y, 'D--b', ms = 10, mfc = 'm') # fmt 'marker|line|color'

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales)
plt.show()
```

```python
plt.bar(products, sales, color='skyblue', width=.5)
plt.title('Sales by Products')
plt.xlabel('Products')
plt.ylabel('Sales')

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
s = np.random.rand(50) * 100

plt.scatter(x, y, s=s, c=s, cmap='viridis')
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(0, 10, 100)
y = np.sin(x)

axes[0].plot(x, y)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(x, y)
axes[1].hist(np.random.randn(1000))

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

x = np.linspace(0, 10, 100)
y = np.sin(x)

axes[0].plot(x, y)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('x axis')
axes[0].set_ylabel('y axis')

axes[1].hist(np.random.randn(1000))
axes[1].set_title('Histogram of 1000 Random Numbers')
axes[1].set_xlabel('x axis')
axes[1].set_ylabel('y axis')

plt.show()
```
